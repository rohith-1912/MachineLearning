#### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 6 * np.random.rand(100, 1) - 3
y = 0.5 *x**2 + 1.5*x + 2+ np.random.randn(100, 1)
# Quadratic Expression Used - y = 0.5x^2+1.5x+2+outliers
```

```python
plt.scatter(x, y, color = 'g')
plt.xlabel("x dataset")
plt.ylabel('y dataset')
plt.show()
```

##### From the above graph we can see that the Features x and y are not Linear. The relationship between them is Non-Linear. So, we can apply the Polynomial Regression for the dataset.

#### Create Train Test Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#### Satndardize the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

#### Lets Implement the Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regression_1 = LinearRegression()
regression_1.fit(x_train, y_train)
```

```python
y_pred_1 = regression_1.predict(x_test)
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred_1)
print("Accuracy of SImple Linear Regression: ", score)
```

```python
#### Lets visuvalize the model
plt.plot(x_train, regression_1.predict(x_train))
plt.scatter(x_train, y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y Dataset")
```

#### Polynomial Regression

```python
#### Lets apply Polynomial Transformations
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly= PolynomialFeatures(degree=2, include_bias=True)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
from sklearn.linear_model import LinearRegression
regression_2 = LinearRegression()
regression_2.fit(x_train_poly, y_train)
y_pred_2 = regression_2.predict(x_test_poly)
```

```python
from sklearn.metrics import r2_score
score_1 = r2_score(y_test, y_pred_2)
print("Accuracy of Polynomial Regression is: ", score_1)
```

```python
print(regression_2.coef_)
```

```python
print(regression_2.intercept_)
```

```python
plt.scatter(x_train, regression_2.predict(x_train_poly))
plt.scatter(x_train, y_train)
```

#### For Degree 3

```python
poly= PolynomialFeatures(degree=3, include_bias=True)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.linear_model import LinearRegression
regression_2 = LinearRegression()
regression_2.fit(x_train_poly, y_train)
y_pred_2 = regression_2.predict(x_test_poly)
```

```python
from sklearn.metrics import r2_score
score_1 = r2_score(y_test, y_pred_2)
print("Accuracy of Polynomial Regression for Degree 3 is: ", score_1)
```

```python
### Prediction of new Data Set
x_new = np.linspace(-3, 3,200).reshape(200, 1)
x_new_poly = poly.transform(x_new)
x_new_poly
```

```python
y_new = regression_2.predict(x_new_poly)
plt.plot(x_new, y_new, "r-", linewidth = 2, label = "New Predictions")
plt.plot(x_train, y_train, "b. ", label = "Training points")
plt.plot(x_test, y_test, "g. ", label = "Testing points")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
