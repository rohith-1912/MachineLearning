### Data Cleaning

```python
###Missing Values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)] #Finds out in any Columns
```

Here we divide the two reigons with the labels as label 0 and label 1.

```python
dataset.loc[:122,"region"]=0
dataset.loc[122:,"region"]=1
df = dataset
```

```python
df.info()
```

```python
df[['region']] = df[['region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.iloc[['122']]
```

```python
#Remove 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[['122']]
```

```python
df.columns
```

```python
#fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

##### Change the required columns as integer datatype

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

##### Changing other columns to float datatype

```python
objects = [features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]= df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
### Lets save the cleaned Dataset.
df.to_csv("Algerian_forest_fires_Cleaned_dataset.csv", index = False) #This makes sure that the index values doen,t come as the seperate columns.
```

### Exploratory Data Analysis

```python
# Drop day, month, year
df_copy = df.drop(['day', 'month', 'year'], axis = 1)
```

```python
df_copy.head()
```

```python
### Encoding the categories in classes.
#df_copy['Classes'] = np.where(df_copy['Classes']=='not fire', 0, 1) ## This doen't change every columns if there are spaces in the values so we can use str.contains to convert all the
#Categorical values of the classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
### Categories in Classes
df_copy['Classes'].value_counts()
```

```python
import matplotlib.pyplot as plt
print(plt.style.available)
```

```python
### Plo6t density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
### Percentage for piechart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
### Plotting Pie Chart
# plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
